scite3: Tertiary source citation providing further details about the attack site.
dbsource: Source of the data, often the name of the mission or organization.

# **Countplot**

````python
##(Chart-1) How many attacks were there in top 15 countries?
plt.figure(figsize=(22, 8))   # Set figure size
sns.countplot(x='country', data=df, order=df['country'].value_counts().index[:15])
plt.xlabel('Country', fontsize=20)
plt.ylabel('Count of Terrorist Attacks', fontsize=15)
plt.title('Top 15 Countries by Number of Terrorist Attacks', fontsize=20)
plt.xticks(rotation=75)
plt.show()
````

##### 1. Why did you pick the specific chart?

## **Correlation heat map**

````python
## (Chart-2) Relationship between the number of wounded and the number of killed people in various attack .
correlation_matrix = df[['nwound', 'nkill']].corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
````

##### 1. Why did you pick the specific chart?

Heat maps are ideal for visualizing complex data patterns and relationships between variables in a matrix or grid format. They use color gradients to represent values, making it easy to identify trends, correlations, and anomalies at a glance.

##### 2. What is/are the insight(s) found from the chart?

As we observed that no of killed people are directly co-related with no of wounded people.
